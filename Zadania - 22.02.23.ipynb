```python
"""
1. Znajdź wszystkie pozycje w tekście, gdzie znajdują się cyfry
"""

import re

tekst = "Dzisiaj mamy 10 dzień miesiąca 03 roku 2011"
wzorzec = "[0-9]+"
re.findall(wzorzec, tekst)
```

```python
"""
2. Znajdź w tekście łańcuchy, w których liczba znaków jest większa niż 3.
"""

tekst = "Dzisiaj mamy 10 dzień miesiąca 03 roku 2011"
wzorzec = "[A-z0-9ąń]{4,}"
re.findall(wzorzec, tekst)
```

```python
"""
4. Stwórz funkcję, która z wektora łańcuchów będących tweetami wyodrębni wszystkie hasztagi (#tag)
"""
import re

tekst = ['tekst #tag','@ktoś na temat #ha2stag','A tutaj nic','#tylkotag','#pierwszytag cos tam #drugitag']

def hasztag(tweets):
  wynik = [re.findall('#\w+', x) for x in tweets]
  return wynik
  
hasztag(tekst)
```
