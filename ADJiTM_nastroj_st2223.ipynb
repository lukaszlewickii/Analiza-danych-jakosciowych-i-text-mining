## Analiza nastroju z biblioteką `TextBlob`

Metody biblioteki `TextBlob` nie wymagają przygotowywania macierzy term-dokument. Nie jest również wymagane zaawansowane czyszczenie tekstu, gdyż analiza nastrojów sprowadza się do odnalezienie w tekście pewnych "*kluczowych*" słów.

```python
# !pip install textblob
```

```python
from textblob import TextBlob
```

```python
TextBlob(teksty[0]).sentiment
```

```python
{token: TextBlob(token).polarity for token in teksty[3].split()}
```

```python
polar_TextBlob = {'kom_'+str(i+1): TextBlob(kom).polarity for i, kom in enumerate(teksty)} 
print(polar_TextBlob)
```

## Analiza nastroju z biblioteką `nltk`

Na podstawie:
Hutto, C.J. & Gilbert, E.E. (2014). VADER: A Parsimonious Rule-based Model for Sentiment Analysis of Social Media Text. Eighth International Conference on Weblogs and Social Media (ICWSM-14). Ann Arbor, MI, June 2014.

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# import słownika
nltk.download('vader_lexicon')
```

```python
sa_vader = SentimentIntensityAnalyzer()
```

```python
sa_vader.polarity_scores(teksty[0])
```

```python
{token: sa_vader.polarity_scores(token)["compound"] for token in teksty[0].split()}
```

```python
polar_veder = {'kom_'+str(i+1): sa_vader.polarity_scores(kom)["compound"] for i, kom in enumerate(teksty)} 
print(polar_veder)
```
