```
[(..., ...),
(..., ...),
...]
```

Każda krotka zawiera:

  * słownik gdzie:
      * kluczami są wszystkie unikatowe termy występujące w korpusie;
      * wartościami wartości logiczne (`True`/`False`) wskazujące na wystąpienie danego termu w dokumencie;
  * etykietę klasy.

> Przykład:

```
# dla zdań:
'It was great!'
'It is bad'

# obiekt będzie wyglądał
[({'it': True,

   'great': True,
   'is': False,
   'bad': False}, 'Positive),
({'it': True,
   'was': False,
   'great': False,
```
